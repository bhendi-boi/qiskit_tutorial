# Retrieving results from previous runs

```python
from qiskit_ibm_runtime import QiskitRuntimeService
```

```python
file = open("secrets.txt", "r")
api_token = file.read()
file.close()
```

```python
service = QiskitRuntimeService(token=api_token, channel="ibm_quantum")
```

```python
job = service.job("cw7w2jt2802g0081ejb0")
```

```python
result_dist = job.result()[0].data.c.get_counts()
```

```python
from qiskit.visualization import plot_distribution
```

```python
plot_distribution(result_dist)
```
