# Simulate Quantum Circuit using Aer Simulator
-   In this notebook we will simulate the circuit we've created in the last notebook using Aer simulator.

## Library Imports

```python
from qiskit.circuit import QuantumCircuit
from qiskit_aer.primitives import SamplerV2
```

```python
qc = QuantumCircuit(1, 1)  # (no_of_qbits, no_of_cbits)

qc.h(0)

qc.measure(0, 0)

qc.draw("mpl", initial_state=True)
```

```python
sampler = SamplerV2()
job = sampler.run([qc], shots=1024)
```

```python
result_dist = job.result()[0].data.c.get_counts()
```

```python
from qiskit.visualization import plot_distribution
```

```python
plot_distribution(result_dist)
```
